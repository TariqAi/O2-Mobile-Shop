## ðŸ”¹ Explore Data Analysis

```python
# Checking the unique company brand names in the data
data['brand'].unique()
```

```python
# Which brand companies most used
data['brand'].value_counts()
```

```python
data['brand'].value_counts().sort_values(ascending=False)\
.plot(kind='pie', autopct='%1.0f%%', cmap='viridis')
plt.xlabel("Brand names")
plt.ylabel("Count of value")
plt.show()
```

```python
# To visualize the which batterys and rams most used

for i in data[['battery','battery_type', 'ram(GB)', 'resolution', 'os']]:
    data[i].value_counts().sort_values(ascending=False)\
    .plot(kind='bar', figsize=(10,5), color=['#008000', '#17A589', '#000080', '#F1C40F' , '#800000'])
    plt.title(f"Visualize the most used {i} in the data")
    plt.xticks(rotation=90)
    plt.xlabel({i})
    plt.ylabel("Count of values")
    plt.show()
```

```python
#The price column contains the dollar currency, So we convert dollar into Jordan dinar

today_dollar_price = 0.71
data['Jod_price']= data['price(USD)'] * today_dollar_price

sum_of_prices= data.groupby(['brand'])['Jod_price'].sum().sort_values(ascending=False)
print(f'{(sum_of_prices).round(1)}')
```

```python
# Select the categorical columns in the data

for i in data.select_dtypes(include='bool'):
    values=data[i].value_counts()
    plt.figure(figsize=(8,5))
    sns.barplot(data=data, x=values.index, y=values.values)
    plt.title(f"Visualize the most used {i} in the data")
    plt.xlabel(i)
    plt.ylabel("Count of values")
    plt.show()
```

```python
# Top 20 mobile brand with ram wise
brand_ram = pd.DataFrame(data.groupby(['brand'])['ram(GB)'].value_counts().sort_values())
brand_ram.head(30).unstack().style.background_gradient(cmap='coolwarm')
```
